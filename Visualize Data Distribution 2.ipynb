## HW1
use boxplot to compare features between 3 classes

```python
plt.subplot(1,4,1)
features1 = (setosa_np[:,0],versicolor_np[:,0],virginica_np[:,0])
box1 = plt.boxplot(features1)
plt.subplot(1,4,2)
features2 = (setosa_np[:,1],versicolor_np[:,1],virginica_np[:,1])
box2 = plt.boxplot(features2)
plt.subplot(1,4,3)
features3 = (setosa_np[:,2],versicolor_np[:,2],virginica_np[:,2])
box3 = plt.boxplot(features3)
plt.subplot(1,4,4)
features4 = (setosa_np[:,3],versicolor_np[:,3],virginica_np[:,3])
box4 = plt.boxplot(features4)
```

## ใส่ตัวหนังสือ

```python
## ชื่อกราฟ
plt.boxplot(features1)
plt.title ('Feature 1 Comparison');
```

```python
## ใส่ชื่อแต่ละกล่องโดยใช้คำสั่ง label
plt.boxplot(features1,labels=['setosa','versicolor','virginica'])
plt.title ('Feature 1 Comparison');
```

```python
import folium
import pandas as pd
```

```python
data_to_plot = Def_df_dabt
```

```python
data_to_plot
```

```python
country_geo = 'thai.json'
```

```python
new_col = list()
for d in data_to_plot['Region and province']:
    new_col.append(d.lstrip())
data_to_plot['Region and province']=new_col
```

```python
map = folium.Map(location=[10,100],zoom_start=5)
```

```python
map
```

```python
map.choropleth(geo_data=country_geo, data=data_to_plot,
               columns=['Region and province',2549],
               key_on = 'feature.properties.name',
              fill_color='YlGnBu',fill_opacity=0.7,line_opacity=0.2,
              legend_name='Average Debt')
```

```python
map.save('plot_data.html')
```
