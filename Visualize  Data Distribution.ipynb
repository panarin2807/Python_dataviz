# Scatter plot (ดูการกระจายของข้อมูล) 
โดยส่วนมากจากplot 2มิติ


```python
irisdata = pd.read_csv('C:\\Users\\Acer\\Downloads\\iris.data',header=None)
irisdata
```

วิธีปิด output -> cell -> current outputs -> toggle

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np[:5,:] ##ดูตารางย่อๆ
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])
##(เอาทุกแถวเอาหลักที่0 = x),(เอาทุกแถวเอาหลักที่1=y)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=70) 
## s=ขนาดของจุด
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2]) 
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = 'r') 
```

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors =[]
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')       
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = colors) 
```

เขียนโปรแกรมแบบไม่ใส่ค่าcolors

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = setosa_np[:,3], cmap = 'YlOrBr' ) 
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = np.array([[1,0.1,1]])) 
## c = np.array([[1,0.5,0.8]])) =(แดง,เหลือง,น้ำเงิน) -> การผสมสี max=1
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = setosa_np[:,3], cmap = 'YlOrBr' , marker = '*') 
##marker เปลี่ยนจุดให้เป็นรูปต่างๆ
```
